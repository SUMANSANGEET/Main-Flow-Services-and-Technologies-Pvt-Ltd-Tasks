**Objective:
Analyze a dataset of student exam scores and answer specific questions, presenting findingsusing Python libraries and techniques for data analysis.**

**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 1: Load the dataset**

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip"
dataset_path="student-mat.csv"
```

```python
# Download and load the dataset
import urllib.request
import zipfile
```

```python
urllib.request.urlretrieve(url, "student.zip")
```

```python
# Extract the dataset

with zipfile.ZipFile("student.zip","r") as zip_ref:
    zip_ref.extractall(".")
```

```python
# Load the dataset into dataframe

import pandas as pd
data=pd.read_csv("student-mat.csv", sep=";")
print("Data loaded successfully!")
data
```

**Step-2: Data Exploration**

```python
data.head()
```

```python
print("\nDataset Info")
data.info()
```

**Data Cleaning**

```python
# Check for missing values

print("\nMissing Values:")
data.isnull().sum()
```

```python
# Remove Duplicates

data=data.drop_duplicates()
data
```

**Step-4 :Data Analysis**

```python
# Question 1: What is the average score in math (G3)?

avg_score=data['G3'].mean()
print(f"\nAverage Math Score (G3): {avg_score:.2f}")
avg_score
```

```python
# Question 2: How many students scored above 15 in their final grade (G3)?

students_above_15=len(data[data['G3'] > 15])
print(f"Number of students scoring above 15: {students_above_15} ")
students_above_15
```

```python
# Question 3: How many students scored above 15 in their final grade (G3)?

correlation=data['studytime'].corr(data['G3'])
print(f"Correlation between study time and final grade: {correlation:.2f}")
correlation
```

```python
# Question 4: Which gender has a higher average final grade (G3)?

avg_grade_by_gender=data.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender:")
avg_grade_by_gender
```

**Step 5: Data Visualization**

```python
# Histogram of final grades

plt.figure(figsize=(8,5))
plt.hist(data['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot of study time vs. final grade

plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='studytime', y='G3', hue='sex')
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time (hours)")
plt.ylabel("Final Grade")
plt.legend(title="Gender")
plt.show()
```

```python
# Bar chart of average scores by gender

plt.figure(figsize=(8,5))
avg_grade_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title("Average final grade by Gender")
plt.ylabel("Average Final Grade")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.show()
```
